```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from rebinding.fig_utils import defaultStyle, save_fig
defaultStyle(fs=14)
```

```python
data = np.load("../data/BYL155/BYL155_1uM_no_DTX_fl.npz")
fl_mat_0 = data["fl0"]
fl_mat_1 = data["fl1"]

_, ax = plt.subplots(2, 1, figsize=(4, 4))
plt.subplots_adjust(hspace=0.1)
sns.heatmap(fl_mat_0, ax=ax[0], cmap="Reds", vmin=200, vmax=420)
sns.heatmap(fl_mat_1, ax=ax[1], cmap="Reds", vmin=200, vmax=420)
for bx in ax:
    bx.set_yticks(np.arange(0, fl_mat_0.shape[0], 120))
    bx.set_yticklabels(np.arange(0, fl_mat_0.shape[0], 120) // 60)
    bx.set_ylabel("Time (h)")
ax[0].set_xticks([])
ax[1].set_xticks(range(0, 14, 2))
ax[1].set_xticklabels([50 * i for i in range(0, 14, 2)])
ax[1].set_xlabel("Distance (μm)")
save_fig("SirTub_efflux_heatmap", fmt="png", dpi=300)
```

```python
data = np.load("../data/BYL155/BYL155_1uM_no_DTX_fl.npz")
data = np.load("../data/BYL215/BYL215_no_10uM_DTX_2_fl.npz")
fl_mat_0 = data["fl0"]
fl_mat_1 = data["fl1"]
fl_mat_0_max = np.percentile(fl_mat_0, 99.9)
fl_mat_1_max = np.percentile(fl_mat_1, 99.9)
fl_mat_0 /= fl_mat_0_max
fl_mat_1 /= fl_mat_1_max
_, ax = plt.subplots(2, 1, figsize=(4, 4))
plt.subplots_adjust(hspace=0.1)
sns.heatmap(fl_mat_0, ax=ax[0], cmap="Reds")
sns.heatmap(fl_mat_1, ax=ax[1], cmap="Reds")
for bx in ax:
    bx.set_yticks(np.arange(0, fl_mat_0.shape[0], 120))
    bx.set_yticklabels(np.arange(0, fl_mat_0.shape[0], 120) // 60)
    bx.set_ylabel("Time (h)")
ax[0].set_xticks([])
ax[1].set_xticks(range(0, 13, 2))
ax[1].set_xticklabels([50 * i for i in range(0, 13, 2)])
ax[1].set_xlabel("Distance (μm)")
save_fig("SirTub_uptake_heatmap", fmt="png", dpi=300)
```
