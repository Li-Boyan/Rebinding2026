```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.integrate import solve_ivp

from rebinding.efflux_model_collection import washout_zero_bound_extra_compartment
from rebinding.fig_utils import defaultStyle, save_fig
```

# Load data and presettings

```python
defaultStyle(fs=14)
params_df_with_bs = pd.read_csv("num_expt/params_df_with_bs.csv")
drugs = [
    "PTX",
    "DTX",
    "Rapamycin",
    "Trametinib",
    "Selumetinib",
    "Colchicine",
    "CA4",
    "VB",
    "Ixazomib",
    "Alisertib",
    "Ispinesib",
    "Erlotinib",
    "Topotecan",
]
set3_colors = mpl.colormaps["Set3"]
colors = [
    set3_colors(0),
    set3_colors(0),
    set3_colors(9),
    set3_colors(2),
    set3_colors(2),
    set3_colors(3),
    set3_colors(3),
    set3_colors(3),
    set3_colors(4),
    set3_colors(5),
    set3_colors(5),
    set3_colors(6),
    set3_colors(7),
]
```

# Quantify compartments

```python
compartment_df = (
    params_df_with_bs[["Drug", "f", "kd_ratios_bs", "m", "kd_bs", "mu_bs"]]
    .groupby("Drug")
    .median()
)
compartment_df = compartment_df.reindex(drugs)
```

```python
factor = 10
c = compartment_df.kd_bs * factor
# c = 0.5
compartment_df["xc"] = c * compartment_df.f / (1 + compartment_df.f)
compartment_df["nsc"] = c
compartment_df["tc"] = c / (compartment_df.kd_bs + c) * compartment_df.m
compartment_df["total_compart"] = (
    compartment_df.xc + compartment_df.tc + compartment_df.nsc
)
for c in ["tc", "xc", "nsc"]:
    compartment_df[c + "_norm"] = compartment_df[c] / compartment_df.total_compart
compartment_df

_, ax = plt.subplots(figsize=(7, 4))
ax.barh(
    compartment_df.index,
    compartment_df.xc_norm,
    color="orange",
    alpha=0.9,
    label="Drug trap",
)
ax.barh(
    compartment_df.index,
    compartment_df.tc_norm,
    left=compartment_df.xc_norm,
    color="blue",
    alpha=0.6,
    label="Target",
)
ax.barh(
    compartment_df.index,
    compartment_df.nsc_norm,
    left=compartment_df.xc_norm + compartment_df.tc_norm,
    color="green",
    alpha=0.6,
    label="Nonspecific/free",
)
ax.set_yticklabels(drugs)
ax.set_xlim([0, 1])
# ax.set_title("500nM nonspecific binding + free drugs", fontsize=14, fontweight="bold")
ax.invert_yaxis()
ax.legend(bbox_to_anchor=(0.0, 1.05, 0.9, 0.1), fontsize=12, ncols=3)
save_fig("fig4-compartment", fmt="png", dpi=300)
```

# Quantify target retention contributions

```python
med_df = (
    params_df_with_bs[
        ["Drug", "kon_bs", "koff_bs", "m_bs", "mu_bs", "k1", "k2"]
    ]
    .groupby("Drug")
    .median()
)
med_df = med_df.reindex(drugs)
med_df["t_res_bs"] = np.log(2) / med_df.koff_bs
med_df
```

```python
def residence_time(kon, koff, m, mu, k1, k2, no_rebinding=False):
    f = k1 / k2
    kd = koff / kon
    kon = 0 if no_rebinding else kon
    y0 = (10 * kd + m, m, 10 * kd * f / (1 + f))
    run_time = 0
    while y0[1] > m * 0.5:
        sol = solve_ivp(
            washout_zero_bound_extra_compartment,
            y0=y0,
            t_span=[0, 8],
            args=[kon, koff, m, mu, k1, k2],
            max_step=0.1,
        )
        y0 = sol.y[:, -1]
        run_time += 1
    try:
        tau = sol.t[np.where(sol.y[1] <= 0.5 * m)[0][0]]
    except:
        print(sol.y, m)
    return (run_time - 1) * 8 + tau


med_df["t_res_rebind"] = med_df.apply(
    lambda df: residence_time(df.kon_bs, df.koff_bs, df.m_bs, df.mu_bs, df.k1, df.k2),
    axis=1,
)
med_df["t_res"] = med_df.apply(
    lambda df: residence_time(df.kon_bs, df.koff_bs, df.m_bs, df.mu_bs, df.k1, df.k2, no_rebinding=True),
    axis=1,
)
```

```python
med_df["t_rebind_x_norm"] = (
    med_df.t_res_rebind - med_df.t_res_bs
) / med_df.t_res_rebind
med_df["t_res_norm"] = med_df.t_res / med_df.t_res_rebind
med_df
```

```python
_, ax = plt.subplots(figsize=(6, 4))
ax.barh(
    med_df.index,
    med_df.t_res_norm,
    color="green",
    alpha=0.8,
    label="Dissociation",
)
ax.barh(
    med_df.index,
    med_df.t_rebind_x_norm,
    left=med_df.t_res_norm,
    color="blue",
    alpha=0.4,
    label="Rebinding",
)
# ax.set_title(r"10$K_d$ nonspecifc binding + free drugs", fontsize=14, fontweight="bold")
ax.legend(bbox_to_anchor=(0.15, 1.05, 0.7, 0.1), fontsize=12, ncols=2)
ax.set_xlim([0, 1])
ax.invert_yaxis()
save_fig("fig4-off-contributions", fmt="png", dpi=300)
```

# Quantify total retention contributions

```python
med_df = (
    params_df_with_bs[
        ["Drug", "kon_bs", "koff_bs", "m_bs", "mu_bs"]
    ]
    .groupby("Drug")
    .median()
)
med_df = med_df.reindex(drugs)
med_df["t_res"] = np.log(2) / med_df.koff_bs
med_df["koff_eff"] = med_df.apply(lambda df: df.mu_bs / (df.mu_bs + df.kon_bs * df.m_bs * 0.5) * df.koff_bs, axis=1)
med_df["t_res_eff_x"] = np.log(2) / med_df.koff_eff - med_df.t_res
med_df["t_efflux"] = np.log(2) / med_df.mu_bs
med_df["t_total"] = med_df.t_res_eff_x + med_df.t_res + med_df.t_efflux
for t in ["t_res", "t_res_eff_x", "t_efflux"]:
    med_df[t + "_norm"] = med_df[t] / med_df.t_total
med_df
```

```python
_, ax = plt.subplots(figsize=(7, 4))
ax.barh(
    med_df.index,
    med_df.t_res_norm,
    color="green",
    alpha=0.8,
    label="Dissociation",
)
ax.barh(
    med_df.index,
    med_df.t_res_eff_x_norm,
    left=med_df.t_res_norm,
    color="blue",
    alpha=0.4,
    label="Rebinding",
)
ax.barh(
    med_df.index,
    med_df.t_efflux_norm,
    left=med_df.t_res_norm + med_df.t_res_eff_x_norm,
    color="orange",
    alpha=0.4,
    label="Diffusion",
)
# ax.set_title(r"10$K_d$ nonspecifc binding + free drugs", fontsize, ncols=2=14, fontweight="bold")
ax.legend(bbox_to_anchor=(0.0, 1.05, 1, 0.1), fontsize=12, ncols=3)
ax.set_xlim([0, 1])
ax.invert_yaxis()
save_fig("fig4-efflux-contributions", fmt="png", dpi=300)
```
